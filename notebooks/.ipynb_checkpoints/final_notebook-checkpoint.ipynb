## How to Run

To run the manuscript, please first download all the trajectories from the Zenodo repository (https://doi.org/10.5281/zenodo.17315520) and place them in the scr folder in the main directory of the repository.

- **Python version:** 3.9+  
- **Environment:** Conda
- **Dependencies:** numpy, MDanalysis, seaborn, matplotlib

## Import and libraries

```python
import MDAnalysis as mda
import numpy as np
from numpy.linalg import norm
import seaborn as sns
import matplotlib.pyplot as plt
from MDAnalysis.lib.distances import distance_array
```

```python
dict_chains_to_rp={'only':'',
                   'BO': 'RP.093',
                   'JK':'RP.057',
                   'LM':'RP.035',
                   'HI':'RP.047',
                   'DE':'RP.052',
                   'LM_BO':'RP.035+RP.093',
                   'LM_HI':'RP.035+RP.047',
                   'LM_JK':'RP.035+RP.057',
                   'HI_JK':'RP.047+RP.057',
                   'LM_HI_JK':'RP.035+RP.047+RP.057',
                   'HI_JK_DE':'RP.052+RP.047+RP.057'}
```

```python
d0s_averages={}
d0s_stdev={}
```

## RIPR all

```python
u = mda.Universe("../src/final_sims/ripr_all_domain/RIPR/top.pdb", "../src/final_sims/ripr_all_domain/RIPR/RIPR_aligned.dcd")

domain1 = u.select_atoms("name CA and resid 1-615")     # Adjust selection
domain2 = u.select_atoms("name CA and resid 861-953")
all_atoms = u.atoms

# Store results
d0s = []

for ts in u.trajectory:
    # align COM1 to reference

    com1 = domain1.center_of_mass()
    com2 = domain2.center_of_mass()
    vec = com2 - com1                    # vector from COM1 → COM2
    
    d0 = norm(vec)
    d0s.append(d0)

# --- Output (example print) ---
```

```python
d0s_means=[]
for argmax in range(0,len(d0s),1000):
    d0s_means.append(np.median(d0s[0:argmax]))

fig,ax=plt.subplots(1,1,figsize=(4, 3))
x=[i*1000 for i in range(len(d0s_means))]
ax.scatter(x,d0s_means)
ax.set_xlabel('steps')
ax.set_ylabel(r'$distance_{RIPR: body-CTD}$ [$\AA$]')
ax.grid(alpha=0.2,zorder=-20)
plt.savefig('../figures/cumulative_ripr_all_COMs_dist.pdf')
```

```python
sns.set(style="whitegrid")
fig,ax=plt.subplots(1,1,figsize=(4.5, 2))
kdedata=sns.kdeplot(d0s[50000:], bw_adjust=0.5, color='teal',ax=ax)
sns.kdeplot(d0s[50000:], fill=True, bw_adjust=0.5, color='teal',ax=ax)
d0s_mean=np.median(d0s[50000:])
d0s_averages['all']=d0s_mean
d0s_stdev['all']=np.std(d0s[50000:])
# extract x and y from the plot line
_,y = kdedata.get_lines()[0].get_data()
# get the max y value and corresponding x
max_y = y.max()
ax.plot([d0s_mean,d0s_mean],[0,max_y*1.10], c='teal')
ax.set_yticks([])
ax.set_ylabel('')
ax.grid(alpha=0.2,zorder=-20)
ax.set_xlabel(r'$distance_{RIPR: body-CTD}$ [$\AA$]')
plt.tight_layout()
ax.set_xlim(20,180)
plt.savefig('../figures/ripr_all_COMs_dist.pdf')
```

## contact maps

```python
start_idr=615
```

```python
cmaps={}
sys_to_eval=['only','BO', 'JK', 'LM_domain', 'HI', 'LM_BO', 'LM_HI_domain', 'LM_JK_domain', 'HI_JK','LM_HI_JK_domain']
sys_to_eval=['only_domain', 'BO_domain','DE_domain', 'LM_domain','JK_domain', 'HI_domain', 'LM_JK_domain', 'HI_JK_domain','LM_HI_JK_domain','HI_JK_DE_domain']
```

```python
for sys_name in sys_to_eval:
    print(sys_name)
    u = mda.Universe(f"../src/final_sims/ripr_{sys_name}/RIPR/top.pdb", f"../src/final_sims/ripr_{sys_name}/RIPR/RIPR_aligned.dcd")
    # Select all beads in the chain
    linker = u.select_atoms("name CA and resid 615-861")  # or a more appropriate selection
    
    n_atoms = len(linker)
    n_frames = len(u.trajectory)
    
    # Contact cutoff (in Å)
    cutoff = 8.0
    
    # Initialize contact frequency matrix
    contact_map = np.zeros((n_atoms, n_atoms))
    
    # Accumulate contacts across frames
    for ts in u.trajectory:
        dists = distance_array(linker.positions, linker.positions)
        contact_matrix = (dists < cutoff).astype(float)
        np.fill_diagonal(contact_matrix, 0)  # ignore self-contacts
        contact_map += contact_matrix
    
    # Normalize to get contact frequency (0–1)
    if sys_name.endswith('_domain'):
        cmaps[sys_name[:-7]]=contact_map/n_frames
    else:    
        cmaps[sys_name]=contact_map/n_frames
```

```python
sys_cmap=['only', 'JK', 'LM', 'HI', 'DE']

fig, axs = plt.subplots(1, len(sys_cmap), figsize=(28, 5), 
                        gridspec_kw={'hspace': 0.2, 'wspace': 0.2})
axs = axs.flatten()

xticks = [i*50 for i in range(5)]
xtick_labels = [start_idr+i*50 for i in range(5)]

for idx_ax, sys_key in enumerate(sys_cmap):
    threshold = 0.15
    
    # difference map
    if sys_key=='only':
        diff = cmaps[sys_key]
        mask = np.abs(diff) >= 0.25

        coords = np.argwhere(mask)       # (i, j) indices
        values = diff[mask]              # differences
        
        # separate positive and negative
        pos_mask = values != 0

        print(np.log(np.abs(values[pos_mask])))
        
        # plot positives (blue)
        axs[idx_ax].scatter(coords[pos_mask, 0], coords[pos_mask, 1],
                            c='black', s=3,
                            alpha=0.8, edgecolors='k', linewidths=0.3,
                            label="positive")
    else:    
        diff = cmaps[sys_key] - cmaps['only']
    
        # keep only points above threshold
        mask = np.abs(diff) >= threshold
    
        coords = np.argwhere(mask)       # (i, j) indices
        values = diff[mask]              # differences
        
        # separate positive and negative
        pos_mask = values > 0
        neg_mask = values < 0
        
        # plot positives (blue)
        axs[idx_ax].scatter(coords[pos_mask, 0], coords[pos_mask, 1],
                            c='blue', s=np.abs(values[pos_mask])*100,
                            alpha=0.8, edgecolors='k', linewidths=0.3,
                            label="positive")
        
        # plot negatives (red)
        axs[idx_ax].scatter(coords[neg_mask, 0], coords[neg_mask, 1],
                            c='red', s=np.abs(values[neg_mask])*100,
                            alpha=0.8, edgecolors='k', linewidths=0.3,
                            label="negative")
    
    axs[idx_ax].set_title(f"RIPR - {dict_chains_to_rp[sys_key]}", fontsize=16)
    axs[idx_ax].set_xlabel('residue', fontsize=16)
    axs[idx_ax].set_xticks(xticks)
    axs[idx_ax].set_xticklabels(xtick_labels, fontsize=16)
    axs[idx_ax].set_yticks(xticks)
    axs[idx_ax].set_yticklabels(xtick_labels, fontsize=16)
    axs[idx_ax].grid(axis='both', color='gray', alpha=0.1)

axs[0].set_ylabel('residue', fontsize=16)

# optional: add legend once
axs[0].legend(loc="upper right")

# single colorbar
#fig.colorbar(sc, ax=axs, orientation='vertical', shrink=0.7, label="Δ contact")

plt.savefig('../figures/ripr_cmaps_single_FAb.pdf')
```

```python
sys_cmap=['LM_JK', 'HI_JK', 'LM_HI_JK','HI_JK_DE']

fig, axs = plt.subplots(2, len(sys_cmap), figsize=(27, 13), 
                        gridspec_kw={'hspace': 0.2, 'wspace': 0.2})
axs = np.atleast_2d(axs)

xticks = [i*50 for i in range(5)]
xtick_labels = [start_idr+i*50 for i in range(5)]

threshold = 0.15
idx_col = 0

for idx, sys_key in enumerate(sys_cmap):
    ### --- First row: direct difference --- ###
    diff = cmaps[sys_key] - cmaps['only']
    mask = np.abs(diff) >= threshold
    coords = np.argwhere(mask)
    values = diff[mask]

    pos_mask = values > 0
    neg_mask = values < 0

    axs[0, idx_col].scatter(coords[pos_mask, 0], coords[pos_mask, 1],
                            c="blue", s=np.abs(values[pos_mask])*100,
                            alpha=0.8, edgecolors='k', linewidths=0.3,
                            label="positive")
    axs[0, idx_col].scatter(coords[neg_mask, 0], coords[neg_mask, 1],
                            c="red", s=np.abs(values[neg_mask])*100,
                            alpha=0.8, edgecolors='k', linewidths=0.3,
                            label="negative")

    axs[0, idx_col].set_title(f"RIPR - {dict_chains_to_rp[sys_key]}", fontsize=16)
    axs[0, idx_col].set_xticks(xticks)
    axs[0, idx_col].set_xticklabels(xtick_labels, fontsize=16)
    axs[0, idx_col].set_yticks(xticks)
    axs[0, idx_col].set_yticklabels(xtick_labels, fontsize=16)
    axs[0, idx_col].grid(axis='both', color='gray', alpha=0.1)

    ### --- Second row: adjusted difference --- ###
    filtered_map = np.where(np.abs(diff) >= threshold, diff, 0.0)
    for elem in sys_key.split('_'):
        cmap_rm = np.where(np.abs(cmaps[elem] - cmaps['only']) >= threshold,
                           cmaps[sys_key] - cmaps['only'], 0.0)
        filtered_map = filtered_map - cmap_rm

    mask2 = np.abs(filtered_map) >= 0
    coords2 = np.argwhere(mask2)
    values2 = filtered_map[mask2]

    pos_mask2 = values2 > 0
    neg_mask2 = values2 < 0

    axs[1, idx_col].scatter(coords2[pos_mask2, 0], coords2[pos_mask2, 1],
                            c="blue", s=np.abs(values2[pos_mask2])*200,
                            alpha=0.8, edgecolors='k', linewidths=0.3)
    axs[1, idx_col].scatter(coords2[neg_mask2, 0], coords2[neg_mask2, 1],
                            c="red", s=np.abs(values2[neg_mask2])*200,
                            alpha=0.8, edgecolors='k', linewidths=0.3)

    axs[1, idx_col].set_title(r"$\Delta$ " + f"[RIPR + FAb] - [FAb]", fontsize=16)
    axs[1, idx_col].set_xlabel("residue", fontsize=16)
    axs[1, idx_col].set_xticks(xticks)
    axs[1, idx_col].set_xticklabels(xtick_labels, fontsize=16)
    axs[1, idx_col].set_yticks(xticks)
    axs[1, idx_col].set_yticklabels(xtick_labels, fontsize=16)
    axs[1, idx_col].grid(axis='both', color='gray', alpha=0.1)

    idx_col += 1

axs[0, 0].set_ylabel("residue", fontsize=16)
axs[1, 0].set_ylabel("residue", fontsize=16)

# optional: one legend only
axs[0, 0].legend(loc="upper right")
plt.savefig('../figures/ripr_cmaps_complexes.pdf')
```
