## Import and functions

```python
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import combinations
from scipy.spatial import ConvexHull

import MDAnalysis as mda
from MDAnalysis.analysis import align, rms
from MDAnalysis.analysis.dihedrals import Dihedral
from MDAnalysis import Merge
import MDAnalysis.analysis.diffusionmap as diffusionmap
from MDAnalysis.coordinates.PDB import PDBWriter
from MDAnalysis.coordinates.memory import MemoryReader
from MDAnalysis.coordinates.DCD import DCDWriter

from sklearn.manifold import TSNE
from sklearn.cluster import HDBSCAN
```

```python
def tsne_subset_projection(distances, fit_range, perplexity=300, random_state=42):
    """
    Fit t-SNE on a subset of frames, then use the subset embedding as initialization
    to run t-SNE on the full dataset.
    
    Parameters:
    - distances: (n_frames, n_pairs) full dataset
    - subset_indices: list or np.array of indices to use for fitting
    """
    subset_data = distances[fit_range[0]:fit_range[1]]

    # First fit t-SNE on the subset
    tsne_subset = TSNE(n_components=2, perplexity=perplexity, random_state=random_state, metric='euclidean')
    subset_embedding = tsne_subset.fit_transform(subset_data)

    # Use subset embedding as initial positions for full t-SNE
    init_embedding = np.random.normal(0, 1e-4, size=(distances.shape[0], 2))
    init_embedding[fit_range[0]:fit_range[1]] = subset_embedding

    # Fit again on full dataset using custom init
    tsne_full = TSNE(n_components=2, perplexity=perplexity, random_state=random_state, 
                     init=init_embedding, metric='euclidean', n_iter=500)
    full_embedding = tsne_full.fit_transform(distances)

    return full_embedding
```

```python
def compute_com_distances(u, residue_ranges,end_range=-1):
    """
    Given an MDAnalysis Universe and residue ranges, compute the pairwise 
    distances between COMs for each frame. Returns a 2D array of shape 
    (n_frames, n_pairs), where each row contains the upper triangle distances 
    (excluding the diagonal) as a flattened 1D array.
    """
    com_groups = [u.select_atoms(f"resid {res[0]}") for res in residue_ranges]
    n_groups = len(com_groups)
    n_frames = len(u.trajectory[0:end_range])
    n_pairs = n_groups * (n_groups - 1) // 2
    distances = np.zeros((n_frames, n_pairs))
    for i, ts in enumerate(u.trajectory[0:end_range]):
        coms = [group.center_of_mass() for group in com_groups]
        # Compute pairwise distances for unique pairs only
        dists = [np.linalg.norm(coms[i] - coms[j]) for i, j in combinations(range(n_groups), 2)]
        distances[i] = dists

    return distances
```

## Comparison trajectories for different RIPR combinations

```python
ref = mda.Universe("../src/final_sims/ripr_only_domain/RIPR.pdb")
# --- Load trajectory ---
u = mda.Universe("../src/final_sims/ripr_only_domain/RIPR/top.pdb", "../src/final_sims/ripr_only_domain/RIPR/RIPR_aligned.dcd")
sys_to_analyse=['only','BO','HI_JK_DE','LM','LM_HI_JK']

all_coms_dists = []
for sys_name in sys_to_analyse:
    u = mda.Universe(f"../src/final_sims/ripr_{sys_name}_domain/RIPR/top.pdb", f"../src/final_sims/ripr_{sys_name}_domain/RIPR/RIPR_aligned.dcd")
    com_dists = compute_com_distances(u, selected_coms)
    all_coms_dists.append(com_dists)
# Concatenate all COMs after loop
all_coms_dists_np = np.concatenate(all_coms_dists, axis=0)
```

```python
all_coms_dists_np_reduced=all_coms_dists_np[::5]
```

```python
len_trajs=[100000,100000,150000,100000,150000]
subsamples_traj=np.array(len_trajs)/5
```

```python
X=tsne_subset_projection(all_coms_dists_np_reduced,(0,120000))
```

```python
fig,(axs)=plt.subplots(1,len(sys_to_analyse),figsize=(30,5))
idx_ax=0

for idx in range(len(sys_to_analyse)):
    start=int(np.sum(subsamples_traj[:idx]))
    end=int(np.sum(subsamples_traj[:(idx+1)]))
    print(start,end)
    # Select the TSNE data chunk for this system
    chunk = X[start:end, :]
    
    # Plot density map
    axs[idx_ax].hexbin(chunk[:, 0], chunk[:, 1], gridsize=20, cmap=cmap1, vmax=300,
                       extent=(np.min(X[:, 0])-0.5, np.max(X[:, 0])+0.5,
                               np.min(X[:, 1])-0.5, np.max(X[:, 1])+0.5))
    
    # Run HDBSCAN clustering
    clusterer = HDBSCAN(min_cluster_size=50)
    labels = clusterer.fit_predict(chunk)
    
    # Overlay cluster boundaries
    unique_labels = np.unique(labels[labels >= 0])  # ignore noise (-1)
    for lab in unique_labels:
        cluster_points = chunk[labels == lab]
        
        # Draw convex hull for each cluster
        if len(cluster_points) >= 3:  # need at least 3 points for a hull
            hull = ConvexHull(cluster_points)
            hull_vertices = np.append(hull.vertices, hull.vertices[0])  # close loop
            axs[idx_ax].plot(cluster_points[hull_vertices, 0],
                             cluster_points[hull_vertices, 1],
                             color='red', linewidth=1.0, alpha=0.8)
    
    # Overlay centroids
    for lab in unique_labels:
        cluster_points = chunk[labels == lab]
        centroid = np.mean(cluster_points, axis=0)
        axs[idx_ax].plot(centroid[0], centroid[1], 'ko', markersize=4, alpha=0.9)
        
    # Styling
    axs[idx_ax].set_xlim(np.min(X[:, 0])-5, np.max(X[:, 0])+5)
    axs[idx_ax].set_ylim(np.min(X[:, 1])-5, np.max(X[:, 1])+5)
    axs[idx_ax].set_yticks([])    
    axs[idx_ax].set_xticks([])
    axs[idx_ax].set_xlabel('t-SNE 1',fontsize=20)
    axs[idx_ax].set_facecolor('#EDF2EA')
    axs[idx_ax].set_title(sys_to_analyse[idx],fontsize=20)

    idx_ax+=1
axs[0].set_ylabel('t-SNE 2',fontsize=20)

plt.savefig('../figures/cluster_all_md_with_circle_clusters.pdf')
```
